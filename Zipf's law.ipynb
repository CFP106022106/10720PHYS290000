```python
fp = open('Alice.txt','r', encoding= 'UTF-8')

#Read the first line from the file
line = fp.readline()
print(line)
print(type(line))

#Read the second line 
line = fp.readline()
print(line)

#Read the third line
line = fp.readline()
print(line)

#Split the line
s = line.split()
print(s)
print(type(s))

#Close the file
fp.close()
```

```python
import matplotlib.pyplot as plt
fp = open('Alice.txt','r', encoding= 'UTF-8')
line = fp.readline()
my_dict = {}

while line:
    s= line.split()
    for x in s:
        if x not in my_dict:
            my_dict[x]= 1
        else:
            my_dict[x] += 1             # my_dict[x] = my_dict[x] + 1
    line = fp.readline()
   
fp.close()

num = []
for key in my_dict:
    num.append(my_dict[key])
    
print(len(num))
num.sort()
num.reverse()

plt.loglog(range(len(num)),num)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.show

```
